```python
# how many employees left and how many didn't
left_count = data['left'].value_counts()
# print(left_count)

# visualizing the comparison
# Use seaborn's countplot to visualize the comparison
plt.figure(figsize=(5, 4))
sns.countplot(x='left', data=data, palette=['skyblue', 'salmon'])

# Add a title and labels for clarity
plt.title('Comparison of Employees Who Left vs. Stayed', fontsize=16)
plt.xlabel('Employee Status (0 = Stayed, 1 = Left)', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)

# Optional: To show the ratio as a pie chart
counts = data['left'].value_counts()
plt.figure(figsize=(4, 4))
plt.pie(counts, labels=['Stayed (0)', 'Left (1)'], autopct='%1.1f%%', startangle=0, colors=['skyblue', 'salmon'])
plt.title('Proportion of Employees Who Left vs. Stayed', fontsize=16)
plt.ylabel('') # Hides the 'left' label on the y-axis
```
